```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
total_players=len(purchase_data["SN"].unique())
tp_df=pd.DataFrame([total_players],columns=["Total Amount of Players"])
tp_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=len(purchase_data["Item ID"].unique())

price_sum=purchase_data["Price"].sum()
price_sum_str="${:,.2f}".format(price_sum)

average_price=(price_sum/len(purchase_data["Price"]))
average_price_str="${:,.2f}".format(average_price)

total_purchases=purchase_data["Price"].count()

summary_df=pd.DataFrame([[unique_items,average_price_str,total_purchases,price_sum_str]],
                        columns=["Amount of Unique Items","Average Price","Amount of Purchases","Total Spent"])
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data_gender=purchase_data.drop_duplicates(subset=["SN"])

gender_grp=purchase_data_gender.groupby("Gender")
gender_df=pd.DataFrame(gender_grp["Gender"].count())
gender_df=gender_df.rename(columns={"Gender":"Total Count"})
gender_df

per_cent_gender_df=pd.DataFrame(gender_grp["Gender"].count()/total_players)*100
per_cent_gender_df=per_cent_gender_df.rename(columns={"Gender":"Percentage of Players"})
per_cent_gender_df=per_cent_gender_df["Percentage of Players"].map("{:,.2f}%".format)
per_cent_gender_df

gender_demo_df=pd.merge(gender_df,per_cent_gender_df, on="Gender")
gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# PURCHASE COUNT
purchase_analysis=purchase_data.groupby("Gender")
purchase_count=pd.DataFrame(purchase_analysis["Purchase ID"].count())
purchase_count=purchase_count.rename(columns={"Purchase ID":"Purchase Count"})
purchase_count

# initial sum
sum_analysis=purchase_analysis["Price"].sum()
sum_purchase=pd.DataFrame(sum_analysis)
sum_purchase=sum_purchase.rename(columns={"Price":"Total Purchase Value"})
sum_purchase_str=sum_purchase

# AVERAGE PURCHASE PRICE
purchase_average=pd.DataFrame(sum_purchase["Total Purchase Value"]/purchase_count["Purchase Count"])
purchase_average=purchase_average.rename(columns={0:"Average Purchase Price"})
purchase_average["Average Purchase Price"]=purchase_average["Average Purchase Price"].map("${:,.2f}".format)
purchase_average

# AVG TOTAL PER PERSON
per_person_average=pd.DataFrame(sum_purchase["Total Purchase Value"]/gender_demo_df["Total Count"])
per_person_average=per_person_average.rename(columns={0:"Avg Total Purchase per Person"})
per_person_average["Avg Total Purchase per Person"]=per_person_average["Avg Total Purchase per Person"].map("${:,.2f}".format)
per_person_average

# TOTAL PURCHASE VALUE
sum_purchase_str["Total Purchase Value"]=sum_purchase_str["Total Purchase Value"].map("${:,.2f}".format)
sum_purchase

# MERGE
purchase_anal_df_1=pd.merge(purchase_count, purchase_average, on="Gender")
purchase_anal_df_1.head()
purchase_anal_df_2=pd.merge(sum_purchase, per_person_average, on="Gender")
purchase_anal_df_2
purchase_anal_df_1.merge(purchase_anal_df_2, on="Gender")

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_demo_group=purchase_data_gender
age_demo_group

bins=[0,10,15,20,25,30,35,40,101]
bin_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

age_demo_group["Age Groups"]=pd.cut(age_demo_group["Age"], bins, right=False,labels=bin_labels)
age_demo_group

age_demo_group=age_demo_group.groupby("Age Groups").count()
age_demo_count=pd.DataFrame(age_demo_group["SN"])
age_demo_count=age_demo_count.rename(columns={"SN":"Total Count"})
age_demo_count["Percentage of Players"]=(age_demo_count["Total Count"]/total_players)*100
age_demo_count["Percentage of Players"]=age_demo_count["Percentage of Players"].map("{:,.2f}%".format)
age_demo_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_analysis=purchase_data
age_analysis["Age Ranges"]=pd.cut(age_analysis["Age"], bins,right=False,labels=bin_labels)
age_analysis_1=age_analysis.groupby("Age Ranges").count()
age_anal_df=pd.DataFrame(age_analysis_1["SN"])
age_anal_df=age_anal_df.rename(columns={"SN":"Purchase Count"})
age_anal_df

age_analysis_2=age_analysis.groupby("Age Ranges").sum()

age_anal_df["Average Purchase Price"]=age_analysis_2["Price"]/age_anal_df["Purchase Count"].round(2)
age_anal_df["Average Purchase Price"]=age_anal_df["Average Purchase Price"].map("${:,.2f}".format)

age_anal_df["Avg Total Purchase per Person"]=age_analysis_2["Price"]/age_demo_count["Total Count"]
age_anal_df["Avg Total Purchase per Person"]=age_anal_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

age_analysis_2["Price"]=age_analysis_2["Price"].map("${:,.2f}".format)
age_anal_df["Total Purchase Value"]=age_analysis_2["Price"]
age_anal_df
                               
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_analysis=purchase_data.groupby("SN")
spender_analysis_df=pd.DataFrame(spender_analysis["Purchase ID"].count())

spender_analysis_df=spender_analysis_df.rename(columns={"Purchase ID":"Purchase Count"})

spender_analysis_df["Total Purchase Value"]=spender_analysis["Price"].sum()

spender_analysis_df["Average Purchase Price"]=spender_analysis_df["Total Purchase Value"]/spender_analysis_df["Purchase Count"]
spender_analysis_df["Average Purchase Price"]=spender_analysis_df["Average Purchase Price"].map("${:,.2f}".format)

spender_analysis_df=spender_analysis_df.sort_values("Total Purchase Value", ascending=False)
spender_analysis_df["Total Purchase Value"]=spender_analysis_df["Total Purchase Value"].map("${:,.2f}".format)

spender_analysis_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_analysis=purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
popular_items_analysis=popular_items_analysis.groupby(["Item ID","Item Name"])

purchase_count_df=pd.DataFrame(popular_items_analysis["Item ID"].count())
purchase_count_df=purchase_count_df.rename(columns={"Item ID":"Purchase Count"})

purchase_value=pd.DataFrame(popular_items_analysis["Price"].sum())
purchase_value_df=purchase_value.rename(columns={"Price":"Total Purchase Value"})
purchase_value_df["Total Purchase Value"]=purchase_value_df["Total Purchase Value"].map("${:,.2f}".format)

pop_items_df=pd.merge(purchase_count_df, purchase_value_df, on=["Item ID", "Item Name"])

pop_items_df["Price"]=popular_items_analysis["Price"].mean()
pop_items_df["Price"]=pop_items_df["Price"].map("${:,.2f}".format)
pop_items_df=pop_items_df.sort_values("Purchase Count", ascending=False)
pop_items_df.head()
```
